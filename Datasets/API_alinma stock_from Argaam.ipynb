```python
import requests
import json

url = "https://www.argaam.com/ar//charts-stock-data"

headers = {
    "Content-Type": "application/x-www-form-urlencoded",
    "User-Agent": "Mozilla/5.0",
    "X-Requested-With": "XMLHttpRequest",
    "Origin": "https://www.argaam.com",
    "Referer": "https://www.argaam.com/ar/company/companyoverview/marketid/3/companyid/826"
}

payload = {
    "stockDataConfigurationId": 6,
    "marketId": 3,
    "companyId": 826,
    "GICSSectorId": 0,
    "period": "AY",  # يعني آخر كل السنوات
    "isBusinessFrame": "false"
}

response = requests.post(url, data=payload, headers=headers)

if response.status_code == 200:
    data = response.json()
    # Check the structure of the returned data
    # print(json.dumps(data, indent=2)) # Uncomment to inspect the data structure

    # Assuming the stock data is under the 'Data' key, similar to the previous file
    if "Data" in data and isinstance(data["Data"], list):
        for item in data["Data"]: # Iterate over the list of dictionaries under the 'Data' key
            # Now 'item' is a dictionary, so you can access its elements using string keys
            print(f" Date: {item['date']}, High: {item['high']}, Low: {item['low']},  Open: {item['open']},  Close: {item['close']},  Volume: {item['volume']}")
    else:
        print("❌ 'Data' key not found or is not a list in the response.")
else:
    print("❌ Failed to fetch data:", response.status_code)
```

```python
import csv


# أسماء الأعمدة المراد حفظها
keys = ["date", "open", "high", "low", "close", "volume"]

# فتح ملف CSV والكتابة فيه
with open("al_inma_stock_data.csv", "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.DictWriter(f, fieldnames=keys)
    writer.writeheader()
    # Iterate over the list of stock data dictionaries under the "Data" key
    # Ensure data is a dictionary and contains the "Data" key which is a list
    if isinstance(data, dict) and "Data" in data and isinstance(data["Data"], list):
        for item in data["Data"]:
            # For each item (which is a dictionary), create a new dictionary
            # with only the desired keys and their corresponding values
            row_data = {k: item.get(k, None) for k in keys} # Use .get() for safer access in case a key is missing
            writer.writerow(row_data)
    else:
        print("❌ Data is not in the expected format (dictionary with 'Data' key containing a list). Cannot save to CSV.")

print("✅ تم حفظ البيانات في al_inma_stock_data.csv (إن كان التنسيق صحيحاً)")
```
